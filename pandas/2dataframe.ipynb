## Obtener valores de un Dataframe:
para obtener valores se buscan al igual que en las series salvo que se puede buscar por categorias/columnas o por indices/vectores, en ese orden columnas, filas.

Para obtener los valores:
- .loc[filas, columnas]: mediante cadenas/str
- .iloc[filas, columnas]: mediante posiciones/int, recibe la funcion range por si se desea secuencias

si se utiliza ":" trae todos los valores.

```python
df1['edad']
```

```python
df1[:3]
```

```python
df1['edad']['ana']
```

```python
df1['edad'][:3]
```

```python
atributos = ['edad', 'Q1']
indices = ['ana', 'camilo']
df1[atributos]
```

```python
df1.loc[indices, atributos]
```

```python
df1.loc[:, atributos]
```

```python
df1.loc[indices, :]
```

```python
df1.iloc[[1,4], [1,2,5]]
```

```python
df1.iloc[range(1,4), 1]
```

```python
df1.iloc[range(1,4), :]
```

```python
df1.iloc[:, 1]
```

## Trabajar con variables categoricas:

Se puede utilizar los siguientes metodos:
 - .nunique(): devuelve la cantidad de valores no unicos de cada columna.
 - .unique(): devuelve los valores unicos, pero solo funciona para una columna en cada aplicación.
 - .astype(): se puede definir el tipo de dato que incluye una o un conjunto de columnas.
 >Se puede utilizar de varias formas:
 >1. relizando un filtro de las columnas a cambiar y a esas se le asigna el mismofiltro con el metodo __.astype()__, por ejemplo:
 >>*x*[ [ *'filtro'* ] ] = x[ [ *'filtro'* ] ].astype(*'tipo de dato'*)
 >2. realizando el filtro mediante un diccionario:
 >>*x*.astype( { *'Columna' : 'tipo de dato'* } )
 - .value_counts(): devuelve la frecuencia de cada valor unico en el df; para que muestre valores normalizados (%) se utiliza el *kwargs* __Normalize = True__

- pd.get_dummies( *x* [ *'filtro'* ] ): Es otra forma de trabajar con categorias, es transformar los datos en variables dummies, es decir transforma la categoria en un valor numerico que representa un booleano, es decir 1 para True y 0 pra false.
___

Para manejar variables temporales:

- pd.to_datetime( _column_, _error: 'tipo error'_, _format_=_'tipo formato'_)

___

__*Manejo del dataframe:*__

- para agregar estas columnas al df, se utiliza el siguiente codigo:
>*x* [ *'nueva columna'* ] = _codigo de la información a incluir_ 
- list(x): muestra el nombre de las columnas del dataframe como una lista.
- .rename(columns={*'nombre antiguo'* : *'nombre nuevo'*, inplace=True}: cambia el nombre de una columna, el __*kwargs*__ __inplace__ define que se reemplace el dataframe original. 
- .drop([ *'nombre columna'* ], axis=1, inplace=True): elimina la columna especificada.
> El __*kwargs*__ *Axis* define el eje del dataframe, es decir 0 para las filas y 1 para las columnas, es importante definir este __*kwargs*__ debido a que si se renombraron los indices, busca el valor especificado en las filas por defecto.

- .drop([ *'nombre columna'* ], index=[ _valores de indice_ ], inplace=True): Si se desea además de borrar columnas, tambien unas filas definidas no se utiliza el axis, sino _index_ así:

- .copy(deep=True): genera una copia que incluye los indices asociados al df.

- .convert_dtypes(): convierte de los tipos de variables de acuerdo a las mejores posibilidades para los datos presentes.

```python
print(f'Valores unicos de cada columna: \n{df_Meteoritos.nunique()}')
print(f'\nvalores unicos de la columna nametype: {df_Meteoritos["nametype"].unique()}\n')
#print(df_Meteoritos[['nametype','fall']])
df_Meteoritos[['nametype','fall']] = df_Meteoritos[['nametype','fall']].astype('category')
print(f'\nTipos de datos de las primeras 5 columnas del df Meteoritos:\n {df_Meteoritos.dtypes.head(5)}')
```
