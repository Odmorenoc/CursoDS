# Series de tiempo

Para el manejo de series de tiempo, se debe asignar a una columna el tipo de dato datetime. mediante el método __*pd.to_datetime()*__ así:

```python
df = pd.read_csv('covid_19_data.csv')
df.sample(10)
df['ObservationDate'] = pd.to_datetime(df['ObservationDate'])
```

```python
# solo para ver los nombres de las columnas del DF.
list(df)
```

```python
df = df[['ObservationDate',
 'Country/Region',
 'Confirmed',
 'Deaths',
 'Recovered']]
df
```

```python
df_time = df.groupby('ObservationDate').sum()
df_time.head(5)
```

```python
df1 = df_time['Confirmed'].iloc[10:15]
df1
```

```python
df2 = df_time['Deaths'].iloc[12:17]
df2
```

```python
df1 - df2
```

```python
df_time
```

para conocer las diferencias entre un dato y el anterior se puede utilizar el metodo __*.diff()*__

```python
df_time.diff().describe()
```

```python
df_diff = df_time.diff()
df_diff
```

```python
df_time.head(1).to_dict()
```

el metodo __*.fillna()*__ permite ingresar un diccionario en el cual se ingresan los valores que van a reemplazar los valores *NaN* del df.

```python
df_diff = df_diff.fillna({'Confirmed': 555.0,
 'Deaths': 17.0,
 'Recovered': 28.0})
```

```python
#suma acumulada
df_diff.cumsum()
```

```python
df_diff.resample('M').sum()
```

```python
df_diff.resample('M').count()
```

```python
df_diff.resample('M').mean()
```

```python
df_cum = df_time.resample('12h').sum(min_count=1)
df_cum
```

```python
df_cum = df_cum.interpolate()
df_cum
```

```python
df_cum['rate'] = (1 - df_cum['Deaths']/df_cum['Confirmed'])*100
```

```python
df_cum = df_cum.reset_index()
df_cum
```

```python
df_cum.groupby(pd.Grouper(key='ObservationDate', freq = 'M'))[['rate']].mean()
```

```python
sr = df_cum.groupby(pd.Grouper(key='ObservationDate', freq = '1D'))['rate'].mean()
sr.plot()
```

```python
sr.plot()
sr.rolling(window=7).mean().plot()
sr.rolling(window=14).mean().plot()
```

```python
import numpy as np
```

```python
sr.rolling(window=14).apply(lambda x: np.std(x)).plot()
```

```python
df_time = df.groupby(['Country/Region',
            pd.Grouper(key='ObservationDate',freq='1D')]).sum()
df_time
```

```python
df_china = df_time.loc['Mainland China',:]
```

```python
df_china
```

```python
import matplotlib.pyplot as plt
```

```python
df_china.plot(figsize = (10,7), title = 'CODV-19')
plt.xlabel('Date')
plt.ylabel('People')
plt.show()
```

```python
ax = df_china.plot(figsize = (10,7), title = 'CODV-19',
              legend = False,
              style = ['r-','g--','b:*'])
ax.legend(['1','dos','3'])
plt.xlabel('Date')
plt.ylabel('People')
plt.show()
```

```python
df_monthly = df_china.resample('M').max()
df_monthly
```

```python
df_monthly.plot(figsize = (10,7), kind='bar', stacked = True)
```

```python
df_monthly['Traitment'] = df_monthly['Confirmed'] - df_monthly['Deaths']- df_monthly['Recovered']
df_monthly
```

```python
df_monthly[['Deaths', 'Recovered', 'Traitment']].T.plot(figsize = (10,7), kind = 'pie', subplots=True)
```
