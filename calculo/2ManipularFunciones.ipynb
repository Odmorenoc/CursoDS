## Dependencias

```python
import numpy as np
import matplotlib.pyplot as plt 
```

## Desplazamientos verticales y horizontales

Siendo $c$ una constante mayor que cero, entonces la gr√°fica:

- $y=f(x)+c$ se desplaza $c$ unidades hacia arriba.
- $y=f(x)-c$ se desplaza $c$ unidades hacia abajo.
- $y=f(x-c)$ se desplaza $c$ unidades hacia la derecha.
- $y=f(x+c)$ se desplaza $c$ unidades hacia la izquierda.

```python
N = 1000

def a(x):
  return x**2;

c = 4

x = np.linspace(-10,10, num=N)

y = a(x)
y1 = a(x + c)
y2 = a(x - c)
y3 = a(x) + c
y4 = a(x) - c


fig, ax = plt.subplots(2,3)
ax[0,0].plot(x,y)
ax[0,0].axhline(y=0, color='r')
ax[0,0].axvline(x=0, color='r')
ax[0,1].plot(x,y1)
ax[0,1].axhline(y=0, color='r')
ax[0,1].axvline(x=0, color='r')
ax[0,2].plot(x,y2)
ax[0,2].axhline(y=0, color='r')
ax[0,2].axvline(x=0, color='r')
ax[1,0].plot(x,y)
ax[1,0].axhline(y=0, color='r')
ax[1,0].axvline(x=0, color='r')
ax[1,1].plot(x,y3)
ax[1,1].axhline(y=0, color='r')
ax[1,1].axvline(x=0, color='r')
ax[1,2].plot(x,y4)
ax[1,2].axhline(y=0, color='r')
ax[1,2].axvline(x=0, color='r')

fig.tight_layout()
```

## Alargamientos y compresiones

Siendo $c$ una constante mayor que cero, entonces la gr√°fica:

- $y=c \cdot f(x)$ alarga la gr√°fica verticalmente en un factor de $c$.
- $y= \frac{1}{c} \cdot f(x)$ comprime la gr√°fica verticalmente en un factor de $c$.
- $y=f(c \cdot x)$ comprime la gr√°fica horizontelmente en un factor de $c$.
- $y= f(\frac{1}{c} \cdot x )$ alarga la gr√°fica horizontelmente en un factor de $c$.

```python
N = 1000

def b(x):
  return np.sin(x);

c = 1

x = np.linspace(-15,15, num=N)

y = c*b(x)


fig, ax = plt.subplots()
ax.plot(x,y)
ax.grid()
ax.axhline(y=0, color='r')
ax.axvline(x=0, color='r')
```

```python
c = 2

y1 = c * b(x)
y2 = (1/c) * b(x - c)
y3 = b(x*c)
y4 = b(x/c)

fig, ax = plt.subplots(2,3, sharex=True, sharey=True)
ax[0,0].plot(x,y)
ax[0,1].plot(x,y1)
ax[0,2].plot(x,y2)
ax[1,0].plot(x,y)
ax[1,1].plot(x,y3)
ax[1,2].plot(x,y4)
fig.tight_layout()
```

## Reflexiones

- $y=-f(x)$ refleja la gr√°fica respecto al eje x.
- $y=f(-x)$ refleja la gr√°fica respecto al eje y.

```python
N = 1000

def f(x):
  return np.log2(x);

x = np.linspace(-7,10, num=N)

y = f(x)
y1 = -f(x)
y2 = f(-x)

fig, ax = plt.subplots(3)
ax[0].plot(x,y)
ax[0].grid()
ax[0].set_title("Original")
ax[0].axhline(y=0, color='r')
ax[0].axvline(x=0, color='r')

ax[1].plot(x,y1)
ax[1].grid()
ax[1].set_title("Reflejada respecto X")
ax[1].axhline(y=0, color='r')
ax[1].axvline(x=0, color='r')

ax[2].plot(x,y2)
ax[2].grid()
ax[2].set_title("Reflejada respecto Y")
ax[2].axhline(y=0, color='r')
ax[2].axvline(x=0, color='r')
fig.tight_layout()
```
